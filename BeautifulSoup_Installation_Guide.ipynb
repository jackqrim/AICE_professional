## 3. BeautifulSoup 기본 사용법

````python
# 샘플 HTML 문서
html_content = """
<html>
<head>
    <title>샘플 페이지</title>
</head>
<body>
    <h1>웹 스크래핑 예제</h1>
    <p class="intro">BeautifulSoup을 사용한 HTML 파싱</p>
    <ul>
        <li>항목 1</li>
        <li>항목 2</li>
        <li>항목 3</li>
    </ul>
    <div id="content">
        <p>이것은 본문 내용입니다.</p>
        <a href="https://example.com">예제 링크</a>
    </div>
</body>
</html>
"""

# BeautifulSoup 객체 생성
soup = BeautifulSoup(html_content, 'html.parser')

print("HTML 파싱 완료!")
````

````python
# 다양한 요소 추출 예제

# 1. 제목 추출
title = soup.find('title').text
print(f"페이지 제목: {title}")

# 2. h1 태그 내용 추출
h1 = soup.find('h1').text
print(f"H1 내용: {h1}")

# 3. 클래스로 요소 찾기
intro = soup.find('p', class_='intro').text
print(f"소개 문구: {intro}")

# 4. 모든 li 요소 추출
list_items = soup.find_all('li')
print("\n목록 항목들:")
for item in list_items:
    print(f"- {item.text}")

# 5. ID로 요소 찾기
content_div = soup.find('div', id='content')
print(f"\n콘텐츠 영역: {content_div.find('p').text}")

# 6. 링크 추출
link = soup.find('a')
print(f"링크 텍스트: {link.text}")
print(f"링크 URL: {link.get('href')}")
````

## 4. 실제 웹페이지 스크래핑 예제 (선택사항)

**주의:** 실제 웹사이트를 스크래핑할 때는 해당 사이트의 robots.txt와 이용약관을 확인하세요.

````python
# 간단한 웹페이지 스크래핑 예제 (httpbin.org 사용)
try:
    # requests 라이브러리가 없는 경우 설치
    import requests
except ImportError:
    !pip install requests
    import requests

# 테스트용 HTML 페이지 요청
url = "https://httpbin.org/html"
try:
    response = requests.get(url, timeout=10)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # 페이지 제목 출력
        title = soup.find('title')
        if title:
            print(f"페이지 제목: {title.text}")
        
        # h1 태그 찾기
        h1 = soup.find('h1')
        if h1:
            print(f"H1 내용: {h1.text}")
    else:
        print(f"페이지를 가져올 수 없습니다. 상태 코드: {response.status_code}")
except Exception as e:
    print(f"웹 스크래핑 중 오류 발생: {e}")
    print("인터넷 연결을 확인하거나 로컬 HTML로 테스트하세요.")
````
